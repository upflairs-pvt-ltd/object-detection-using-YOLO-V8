```python
#importing yolo from ultralytics
from ultralytics import YOLO
```

```python
#After executing this cell you will get 2 yolo pretrained model v5 and v8 downloaded as well loaded.
yolo_pretrain_yolov5 = YOLO('yolov5xu.pt')
yolo_pretrain_yolov8 = YOLO('yolov8x.pt')
```

```python
# To perform detection using CLI on a image with the help of downloaded model.
# we have already downloaded 2 yolo versions -->  { yolov5xu.pt , yolov8x.pt }

!yolo task=detect mode=predict model='yolov8x.pt' source="dog.jpg"     # Passed single an image as a source
!yolo task=detect mode=predict model='yolov5xu.pt' source="dog.jpg"

!yolo task=detect mode=predict model='yolov5xu.pt' source="C:\Users\Ranjit\Pictures\Camera Roll\WIN_20240124_05_12_33_Pro.mp4"   # Passed a video as a source
!yolo task=detect mode=predict model='yolov5xu.pt' source=0  # To use webcame give source = 0 and run in terminal


# output will be save inside the runs/detect/predict directory.
# Tu use webcame as source, use source=0
```
